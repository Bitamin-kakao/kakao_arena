```python
val_songs = [[]]

for i in range(len(only_song)):
    while (len(val_songs[i]) < 200) and (200 - len(val_songs[i]) > len(only_song[i])):
        val_songs[i].extend(only_song[i])
    
    val_songs[i].extend(random.sample(only_song[i],200 - len(val_songs[i])))
    val_songs.append([])
    #print(i)
    
del val_songs[-1]
```

```python
from multiprocessing import Pool
import time
import tqdm

def do_work(k):
    ys_list = []
    dist = []
    i=0
    for j in range(len(latent_space)):
        dist.extend([np.linalg.norm(val_latent_space[k]-latent_space[j])])
    order_list = np.array(dist).argsort()

    while len(ys_list)<100:
        if 100 - len(ys_list) >= len(train_genre["songs"][order_list[i]]) :
            for z in range(len(train_genre["songs"][order_list[i]])):
                if train_genre["songs"][order_list[i]][z] not in only_song_raw_df[k][4]:
                    ys_list.append(train_genre["songs"][order_list[i]][z])
        else:
            tmp_list = random.sample(train_genre["songs"][order_list[i]],100 - len(ys_list))
            for z in range(len(tmp_list)):
                if tmp_list[z] not in only_song_raw_df[k][4]:
                    ys_list.append(tmp_list[z])
        i = i+1

        if k%1000==0:
            time.sleep(2)
        
        ys_list = list(set(ys_list))
    
    return ys_list

            
if __name__ == '__main__':
    start_time = time.time()
    with Pool(processes=15) as p:
        result= list(tqdm.tqdm(p.imap(do_work, range(0,len(only_song_raw_df))),total=len(only_song_raw_df)))
    print("--- %s seconds ---" % (time.time()-start_time))     
```
