### 한글자 처리 TF-IDF

**한글자만 다 불러와**

```python
import pandas as pd
from math import log

tmp_list = []
aa = []
for i in finals:
    for j in range(len(i)):
        aa.append(len(i[j])==1)
    if any(aa) :
        tmp_list.append(i)
        aa = [] 
            
tmp_list
```

### 유니크한 한글자 단어들이 들어있는 리스트들

```python
vocabs = list(set([y for x in tmp_list for y in x]))
vocabs
```

### 유니크한 한글자들만

```python
tmp = []
for i in vocabs:
    if len(i) == 1:
        tmp.append(i)
tmp        

N = len(tmp_list)
```

```python
def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in tmp_list:
        df += t in doc
    return log(N/(df+1))

def tfidf(t, d):
    return tf(t,d)*idf(t)
```

```python
result = []
for i in range(N):
    result.append([])
    d = tmp_list[i]
    for j in range(len(tmp)):
        t=tmp[j]
    #for j in range(len(vocab)):
    #    t = vocab[j]
        result[-1].append(tf(t,d))

tf_ = pd.DataFrame(result, columns = tmp)        
#tf_ = pd.DataFrame(result, columns = vocab)
tf_
```
