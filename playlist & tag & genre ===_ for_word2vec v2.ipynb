### 영어 한글자 제거

### TFIDF 전처리
- morphs에서 한 글자인 것만 추출
- sklearn TFIDF가 문장 형태만 인식하기 때문에 join 사용해서 문장처럼 붙여줌 -> tmpp_list

```python
import pandas as pd
import numpy as np

morphs = finals.copy()
tmp_list = []
for j in range(0,len(morphs)):
    idx = np.where(np.array([len(i) for i in morphs[j]])==1)
    if np.array(idx).shape[1]==0:
        continue
    else:
        tmp_list.append(np.array(morphs[j])[idx])
```

```python
# corpus 만들기 위해서 문장처럼 묶기(어짜피 빈칸 있으면 다른 단어로 알아서 인식함)
tmpp_list = []
for i in range(0,len(tmp_list)):
    tmpp_list.append(' '.join(tmp_list[i]))
```

```python
tmpp_list
```

### TFIDF 만들기
- TFIDF는 한글자를 고려하지 않도록 되어있음
- 하지만, 우리는 한글자만으로 TFIDF를 만들어야 하기 때문에 **tocken_pattern**을 넣어주었음

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(token_pattern = r"(?u)\b\w+\b")
A_tfidf_sp = tfidf.fit_transform(tmpp_list)
tfidf_dict = tfidf.get_feature_names()

data_array = A_tfidf_sp.toarray()
data = pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```

```python
tfidf = data[data>0.97].dropna(how='all',axis=1).copy()
```

**포함되어 있지 않은 글자들**

```python
non_in_list= []
for i in data.columns:
    if i in tfidf.columns:
        pass
    else:
        non_in_list.append(i)
non_in_list
```

**선정된 글자들 보기**

```python
[print(x) for x in tfidf.columns]
```

**선정된 한글자들만 남기고 제거**

```python
for i in range(0,len(finals)):
    for j in range(0,len(finals[i])):
        if len(finals[i][j]) == 1:
            if finals[i][j] not in tfidf:
                finals[i][j] =''

for i in range(0,len(finals)):
    while '' in finals[i]:
        finals[i].remove('')
finals   
```

### 합친 파일 저장

```python
i=0
text = open('data/for_word2vec.txt','w')
while i < len(finals):
    for j in range(0,len(finals[i])):
        text.write(finals[i][j])
        if j < len(finals[i])-1:
            text.write(',')
        else: continue
    text.write('\n')
    i += 1
text.close()
```
