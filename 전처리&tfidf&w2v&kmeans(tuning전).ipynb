### 영어 한글자 제거

### TFIDF 전처리
- morphs에서 한 글자인 것만 추출
- sklearn TFIDF가 문장 형태만 인식하기 때문에 join 사용해서 문장처럼 붙여줌 -> tmpp_list

```python
import pandas as pd
import numpy as np

morphs = finals.copy()
tmp_list = []
for j in range(0,len(morphs)):
    idx = np.where(np.array([len(i) for i in morphs[j]])==1)
    if np.array(idx).shape[1]==0:
        continue
    else:
        tmp_list.append(np.array(morphs[j])[idx])
```

```python
# corpus 만들기 위해서 문장처럼 묶기(어짜피 빈칸 있으면 다른 단어로 알아서 인식함)
tmpp_list = []
for i in range(0,len(tmp_list)):
    tmpp_list.append(' '.join(tmp_list[i]))
```

```python
tmpp_list
```

### TFIDF 만들기
- TFIDF는 한글자를 고려하지 않도록 되어있음
- 하지만, 우리는 한글자만으로 TFIDF를 만들어야 하기 때문에 **tocken_pattern**을 넣어주었음

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(token_pattern = r"(?u)\b\w+\b")
A_tfidf_sp = tfidf.fit_transform(tmpp_list)
tfidf_dict = tfidf.get_feature_names()

data_array = A_tfidf_sp.toarray()
data = pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```

```python
tfidf = data[data>0.97].dropna(how='all',axis=1).copy()
```

```python
pd.DataFrame(tfidf.columns)
```

- 한글자만 남겨두고 TFIDF 말고, 모든 글자에 대해서 TFIDF 만든뒤에 한글자만 뽑으려고 했더니 메모리 문제로 에러발생

```python
from sklearn.cluster import KMeans

word_vectors = model_play.wv.syn0 # 어휘의 feature vector
num_clusters = int(256) # 어휘 크기의 1/5나 평균 5단어
print(num_clusters)
num_clusters = int(num_clusters)
```

```python
kmeans_clustering = KMeans(n_clusters=num_clusters)
idx = kmeans_clustering.fit_predict(word_vectors)

idx = list(idx)
names = model_play.wv.index2word
word_centroid_map = {names[i]: idx[i] for i in range(len(names))}
```

```python
for c in range(num_clusters):
    # 클러스터 번호를 출력
    print("\ncluster {}".format(c))
    
    words = []
    cluster_values = list(word_centroid_map.values())
    for i in range(len(cluster_values)):
        if (cluster_values[i] == c):
            words.append(list(word_centroid_map.keys())[i])            
    print(words)
```
