# word2vec grid search (loss)

```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
from gensim.models.callbacks import CallbackAny2Vec

class callback(CallbackAny2Vec):
    """
    Callback to print loss after each epoch
    """
    def __init__(self):
        self.epoch = 0

    def on_epoch_end(self, model):
        loss = model.get_latest_training_loss()
        if self.epoch == 0:
            print('Loss after epoch {}: {}'.format(self.epoch, loss))
        else:
            print('Loss after epoch {}: {}'.format(self.epoch, loss- self.loss_previous_step))
        self.epoch += 1
        self.loss_previous_step = loss
```

# 흐음,,파라미터를 어떻게 넣어줘야할까?

```python
import itertools
def expandgrid(*itrs):
   product = list(itertools.product(*itrs))
   return {'Var{}'.format(i+1):[x[i] for x in product] for i in range(len(itrs))}

window = [3,4,5,6,7,8,9,10]
alpha = [0.01,0.02,0.03]
iter = [5,10,20,50,100]
parm = expandgrid(window, alpha, iter)
```

```python
parm
```

```python
# word2vec with playlist_title
total_loss = []
for i in range(len(parm['Var1'])):
    model_play = Word2Vec(sentences=morphs,
                     size=256, window=parm['Var1'][i],
                     min_count=2,
                     workers=15,
                     sg=1,
                     alpha=parm['Var2'][i],
                     iter=parm['Var3'][i],
                     compute_loss=True,
                     callbacks=[callback()]
                     )#seed=1234)
    total_loss.append(model_play.get_latest_training_loss())

#model_play.save('play_w2v.model') # 모델 저장
```
