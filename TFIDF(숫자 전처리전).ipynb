```python
import pickle
with open("train_genre.pickle","rb")as fr:
    train_genre = pickle.load(fr)
train_genre.head()

for_w = train_genre[['tags','genre']]
for_w.head()

for_w['plylst_title'] = morphs
for_w.head()

finals=[[]]
for i in range(0,len(for_w)):
    if i == 0:
        finals[0] = for_w.iloc[0,0] + for_w.iloc[0,1] + for_w.iloc[0,2]
    else:
        finals.append(for_w.iloc[i,0]+for_w.iloc[i,1]+for_w.iloc[i,2])
finals

for i in range(0,len(finals)):
    for j in range(0, len(finals[i])):
        finals[i][j] = finals[i][j].lower()
finals

finals[5]

for i in range(0,len(finals)):
    finals[i] = list(set(finals[i]))
finals

finals[0]

i=0
text = open('for_word2vec.txt','w')
while i < len(finals):
    for j in range(0,len(finals[i])):
        text.write(finals[i][j])
        if j < len(finals[i])-1:
            text.write(',')
        else: continue
    text.write('\n')
    i += 1
text.close()
```

### TFIDF 전처리
- morphs에서 한 글자인 것만 추출
- sklearn TFIDF가 문장 형태만 인식하기 때문에 join 사용해서 문장처럼 붙여줌 -> tmpp_list

```python
import pandas as pd
from math import log
import numpy as np

tmp_list = []
for j in range(0,len(morphs)):
    idx = np.where(np.array([len(i) for i in morphs[j]])==1)
    if np.array(idx).shape[1]==0:
        continue
    else:
        tmp_list.append(np.array(morphs[j])[idx])
```

```python
# corpus 만들기 위해서 문장처럼 묶기(어짜피 빈칸 있으면 다른 단어로 알아서 인식함)
tmpp_list = []
for i in range(0,len(tmp_list)):
    tmpp_list.append(' '.join(tmp_list[i]))
```

```python
tmpp_list
```

### TFIDF 만들기
- TFIDF는 한글자를 고려하지 않도록 되어있음
- 하지만, 우리는 한글자만으로 TFIDF를 만들어야 하기 때문에 **tocken_pattern**을 넣어주었음

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(token_pattern = r"(?u)\b\w+\b")
A_tfidf_sp = tfidf.fit_transform(tmpp_list)
tfidf_dict = tfidf.get_feature_names()

data_array = A_tfidf_sp.toarray()
data = pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```
